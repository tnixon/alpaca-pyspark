```python
import datetime as dt
from zoneinfo import ZoneInfo

from alpaca_pyspark.common import build_page_fetcher, fetch_all_pages, build_url
```

```python
tz = ZoneInfo("America/New_York")

headers = {
    "Content-Type": "application/json",
    "APCA-API-KEY-ID": dbutils.secrets.get(scope="tnixon_alpaca_api", key="api_key"),
    "APCA-API-SECRET-KEY": dbutils.secrets.get(scope="tnixon_alpaca_api", key="secret"),
}

data_endpoint = "https://data.alpaca.markets/v2"
hist_bars_path = ["stocks", "bars"]

params = {
    "symbols": "AAPL",
    "timeframe": "1Day",
    "start": dt.datetime(2021, 1, 1, tzinfo=tz).isoformat(),
    "end": dt.datetime(2021, 2, 1, tzinfo=tz).isoformat(),
    "limit": 5
}
```

```python
build_url(data_endpoint, hist_bars_path, params)
```

```python
get_page_fn = build_page_fetcher(data_endpoint, headers, hist_bars_path)
```

```python
for pg in fetch_all_pages(get_page_fn, params):
    print(pg)
    print("----")
```
